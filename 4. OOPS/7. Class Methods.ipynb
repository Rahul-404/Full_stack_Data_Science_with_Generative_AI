Q. what if, i wanted to pass name and email with some alternative method?
what are the options available to me?

--> this is where class methods are going to come into picture

```python
class myskills1:

  def __init__(self, name, email):

    self.name = name
    self.email = email

  @classmethod
  def details(cls, name1, email1):
    return cls(name1, email1)

  def student_details(self):
    print(self.name, self.email)
```

we have created method `details` by passing `cls` as reference pointer, and decorated with a built-in decorator `@classmethod` for any classes available inside python.

Let's see what we can achive with it

```python
my1 = myskills1.details("mohan", "mohan@gmail.com")
```

```python
my1.name
```

```python
my1.email
```

- we are able to do function over loading using this `@classmethod` decorator with `detail(cls, name1, email1)` function
- Directly function over loading is not possible in python
- but here we are able to achive it.
- we are able to over load this `__init__()` method with `@classmethod` under `details()`
- `@classmethod` is just able to decorate the function `details()` able to tell class directly that yes it directly belongs to me
- And with name of class we are able to access it
`myskills.details()`

```python
my1.student_details()
```

inside `@classmethod` details() method

```python
class myskills2:

  mobile_number = 4556788912

  def __init__(self, name, email):

    self.name = name
    self.email = email

  @classmethod
  def change_number(cls, mobile):
    myskills2.mobile_number = mobile


  @classmethod
  def details(cls, name1, email1):
    return cls(name1, email1)

  def student_details(self):
    print(self.name, self.email, myskills2.mobile_number)
```

```python
myskills2.change_number(9865451232)
```

```python
myskills2.mobile_number
```

- with the help of class method we are able to change the `mobile number`
- `@classmethod` means we are binding it towards our class

```python
my2_obj = myskills2('sohil', 'sohil@gmail.com')
```

```python
my2_obj.details()
```

```python
my2_obj.details('sohil', 'sohil@gmail.com')
```

```python
my2_obj.name
```

```python
my2_obj.email
```

- each and every class is able to access the `@classmethod`
- when we create multiple instances of class for each instance there will be the number of `student_details()` , lot of memory will be consumed
- but in case of `@classmethod` it created once in memory, for all instances it will be same and accessable
