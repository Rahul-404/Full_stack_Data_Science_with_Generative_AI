Q. if i have a situation, where i would like to call this `"this is the start of my fun"` and `"this is the end of my fun"` with any function. Let's suppose i am going to create `test1(), test2(), test3() ` may be N number of functions and with all the functions is we have to call this `"this is the start of my fun"` and `"this is the end of my fun"` then how i'll be able to do ?


--> Possible solution, is to write this lines in every functions we are going to create.

--> but there is a alternative by using it will be able to achive it.

will create a decorator function, it will make it happen for every function

````python
def deco(func):
  def inner_deco():
    print("this is the start of my fun")
    func()
    print("this is the end of my fun")
  return inner_deco
````

````python
def test1():
  print(4+5)
````

````python
test1()
````

this not the ouput we wanted


expected output:
```
this is the start of my fun
9
this is the end of my fun
```

- we just added `@deco` now it got wrapped up

````python
@deco
def test1():
  print(4+5)
````

````python
test1()
````

- it is used to eliminate the repetative ness from the each and every function going forward

Q. if i would like to know how much time a function is taking for execution each and every time, to calculate the time complexity

````python
import time

def timer_test(func):
  def timer_test_inner():
    start = time.time()
    func()
    end = time.time()
    print(end-start)
  return timer_test_inner
````

````python
def test2():
  print(45+67)
````

````python
test2()
````

````python
@timer_test
def test2():
  print(45+67)
````

````python
test2()
````

````python
@timer_test
def test3():
  for i in range(100000000):
    pass
````

````python
test3()
````

- it helps in reducing our time on repetative tasks
