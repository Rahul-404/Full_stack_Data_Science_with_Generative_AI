```python
import abc

class myskills:

  @abc.abstractmethod
  def student_details(self):
    pass

  @abc.abstractmethod
  def student_assignment(self):
    pass

  @abc.abstractmethod
  def student_marks(self):
    pass
```

```python
class data_science(myskills):

  def student_details(self):
    return "it will try to return a details of data science masters"

  def student_assignment(self):
    return "it will return a details of student assignment for data science masters"

  def student_marks(self):
    return "it will return a details of student marks for data science masters"
```

```python
class web_dev(myskills):

  def student_details(self):
    return "this will return a details of web dev class"

  def student_marks(self):
    return "this will return a details of marks of web dev class"
```

```python
ds = data_science()
ds.student_details()
```

```python
wb = web_dev()
wb.student_details()
```
