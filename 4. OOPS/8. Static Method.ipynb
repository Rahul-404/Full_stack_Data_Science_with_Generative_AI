Q. if i have to create a function whicih should not depend upon object ?

Why above quesion?
- > if go with dependency, then if i have to create 10 or 10 million such objects, then it will create those 10 million functions in memory , so memory consumption will be very very high.
- let's suppose if i have to pass somethong common to my entire class so again in this way will not be able to access
- cause it is something common to my specific object

```python
class myskills1:
  def student_details(self, name, mail_id, number):
      print(name, mail_id, number)

  @staticmethod
  def mentor_class(list_mentor): # withthis will get only once instance irrespective of RAM
    print(list_mentor)

  def mentor(Self, mentor_list): # with # of instance # of code will occupy RAM
    print(mentor_list)
```

```python
my1 = myskills1()
```

```python
my1.mentor(["rahul", "ajay"])
```

```python
my1.mentor_class(["rahul", "ajay"])
```

```python
myskills1.mentor_class(["rahul", "ajay"])
```

- static method will be available to all the instances, without replicating code in RAM

```python
class myskills1:
  def student_details(self, name, mail_id, number):
      print(name, mail_id, number)

  @staticmethod
  def mentor_mail_id(mail_id):
    print(mail_id)

  @staticmethod
  def mentor_class(list_mentor): # withthis will get only once instance irrespective of RAM
    print(list_mentor)
    myskills1.mentor_mail_id(['ajay@gmail.com', 'rahul@gmail.com'])

  @classmethod
  def class_name(cls, class_name):
    cls.mentor_class(["ajay", "rahul"])


  def mentor(Self, mentor_list): # with # of instance # of code will occupy RAM
    print(mentor_list)
    self.mentor_class(["rahul", "ajay"])
```

- calling `mentor_mail_id` it's a static method

- nested `@statismethod` inside `mentor_class()`

- Also calling `@staticmethod` `mentor_class` inside `@classmethod`

- Also calling `@staticmethod` inside instance method

```python
my1 = myskills1()
```

```python
my1.student_details('mohan', 'mohan@gmail.com', 8978564585)
```
