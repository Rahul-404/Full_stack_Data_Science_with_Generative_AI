*Polymorphism:* the ability of something to have or to be displayed in more than one form. the different forms arise because entities can be assigned different meanings and used in various ways in multiple contexts.

```python
def test (a,b):
  return a+b
```

```python
test(4,5)
```

```python
test("rahul ", "shelke")
```

- the above function is showing polymorphism.
- based on provided data it changing it behavior to wards the output.
- it is having multiple characteristics or states as per the data

```python
test([2,3,4,5,5], [4,5,6,7])
```

- now same function working as list concatination, see multiple characteristics

```python
class data_science:
  def syllabus(self):
    print("this is my method for data science syllabus")
```

```python
class web_dev:
  def syllabus(self):
    print("this is my method for web dev ")
```

```python
def class_parcer(class_obj):
  for i in class_obj:
    i.syllabus()
```

```python
obj_data_science = data_science()
```

```python
obj_web_dev = web_dev()
```

```python
class_obj = [obj_data_science, obj_web_dev]
```

```python
class_parcer(class_obj)
```

Let's try to understand what happened here with the output.

- in 1st iteration it accessed object of data science and used its method syllabus
- in 2nd iteration it accessed object of web dev and used its method syllabus
- we can say here, we have achived polymorphism inside my function
