- we have assigned new values to the variables and we are able to access it

-- This kind of things are dangerous! when we are going to do industry grade codeing

Why ?

- because anyone is able to access our variable
- anyone is able to do some sort of modifications or some sort of assignment operation to a variable.
- which is not allowed and we also don't allow such type of modifications in each and every type of classes and objects that we are going to create
- some times we have to take preventive measures and actions that such type of modifications are not allowed against user
- so this is where the concept of "ENCAPSULATION" comes into picture

- Encapsulation is an idea allows us to prevent any kind of direct access or any kind of direct modifications to program in OOPS programming concept.

```python
class car:

  def __init__(self, year, make, model, speed):
    self.__year = year
    self.__make = make
    self.__model = model
    self.__speed = 0

  def set_speed(self, speed):
    self.__speed = 0 if speed < 0 else speed

  def get_speed(self):
    return self.__speed
```

let's see what kind of significance will be provided by `__` prior to variables

```python
obj_car = car(2021, "toyota", "innova", 12)
```

```python
obj_car.__year
```

- if we use tab will not be able to see those variables, cause by adding `__` prior to variables we made them private so nobody can access those variable using class object.
- it's completely prohebited the access of this variables.

```python
obj_car._car__year
```

- but with the `_car` class name we are able to do it, but only when we know the class name and variable name.
-  if we are unknown user we'll not be able to know class name and it's internal structure.

```python
obj_car._car__year
```

```python
obj_car._car__year = 2023
```

```python
obj_car._car__year
```

- As a creator we know the class name and variable name, eles there is no way to know this information for any user who is not part of creation.

```python
obj_car.set_speed(-1234)
```

```python
obj_car.set_speed(1234)
```

- as user , user is able to set speed uisng `set_speed` function but not able to access it cause the variabel `__speed` is private

```python
obj_car._car__speed
```

- to access the speed , we need to have a another method through which it can return.

```python
obj_car.get_speed()
```

```python
class bank_acount:

  def __init__(self, balance):
    self.__balance = balance

  def deposit(self, amount):
    self.__balance = self.__balance + amount

  def withdraw(self, amount):
    if self.__balance >= amount:
      self.__balance = self.__balance - amount

      return True
    else:
      return False

  def get_balance(self):
    return self.__balance

```

```python
obj_bank_account = bank_acount(1000)
```

```python
obj_bank_account.__balance
```

- due to encapsulation we are unable to access the `__balance` variable.

```python
obj_bank_account.get_balance()
```

- As a user we are able to access the balance

```python
obj_bank_account.deposit(6000)
```

```python
obj_bank_account.get_balance()
```

```python
obj_bank_account.withdraw(10000)
```

```python
obj_bank_account.withdraw(2000)
```

```python
obj_bank_account.get_balance()
```
