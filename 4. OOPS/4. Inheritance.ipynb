Inheritance: idea where a class can
inherit a property of another class

```python
class parent:

  def test_parent(self):
    print("This is my parent class")
```

```python
class child(parent):
  pass
```

```python
child_obj = child()
```

```python
child_obj.test_parent()
```

Here we have achieved in heritance

there are two types of inheritance
- multi-level inheritance
- multiple inheritance

### Multi-Level Inheritance

image.png

```python
class class1:
  def test_class1(self):
    print("this is my class1")
```

```python
class class2(class1):
  def test_class2(self):
    print("this is my class2")
```

```python
class class3(class2):
  def test_class3(self):
    print("this is my class3")
```

```python
obj_class3  = class3()
```

```python
obj_class3.test_class1()
```

```python
obj_class3.test_class2()
```

```python
obj_class3.test_class3()
```

- using object of class3 we can access all the class function because of inheritance
- this is called multi-level inheritance

### Multiple Inheritance

image.png

```python
class class1:
  def test_class1(self):
    print("this is my class 1")
```

```python
class class2:
  def test_class2(self):
    print("this is my class 2")
```

```python
class class3(class1, class2):
  pass
```

```python
obj_class3 = class3()
```

```python
obj_class3.test_class1()
```

```python
obj_class3.test_class2()
```

- As we can see are able to access both separate class through object of class3
- this is mutiple inheritance
