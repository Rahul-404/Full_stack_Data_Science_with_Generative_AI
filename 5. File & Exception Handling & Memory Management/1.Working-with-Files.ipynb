*Mode*

|Character | Meaning |
| --- | --- |
| 'r' | open for reading (default) |
| 'w' | open for waiting, truncating the files |
| 'x' | create a new file and open it for writing |
| 'a' | open for writing, appending to the end of the file if it exisits |
| 'b' | binary mode |
| 't' | text mode (default) |
| '+' | open a disk file for updating (reading and writing) |
| 'U' | universal newline mode (deprecated) |

### mode = 'a' (append)

- now let's try it with the other mode `'a'` append mode

```python
f = open(file='test.txt', mode= 'a') # open same file again
```

```python
f.write(text) # adding text
```

```python
f.close() # make sure that it reflects the written
```

- Now , if you open the file it will contain the text twice.
- it's because it go appended with previous text which was available.
- Means with `'a'` append, it will add the information in existing file, it will never trunckate the infromation from file.
