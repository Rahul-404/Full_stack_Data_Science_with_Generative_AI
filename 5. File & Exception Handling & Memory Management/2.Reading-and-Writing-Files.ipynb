- this is a dictionary and also called as JSON (Java Script Object Notation) file.


```python
data = {
    "name": "rahul",
    "main_id": "rahul@gmail.com",
    "phone": 9135446846,
    "subject": ["data science", "big data", "data analytics"]
}
```

```python
import json
```

- will do read and write operation

```python
with open("test1.json", "w") as f:
  json.dump(data, f)
```

```python
# read opeartion
with open("test1.json", "r") as f:
  data1 = json.load(f)
```

```python
data = [["name", "email_id", "number"],
        ["rahul", "rahul@gmail.com", 4566875465],
        ["ajay", "ajay@gmail.com", 5341684568]
        ]
```

```python
import csv
```

### Write CSV

```python
with open("test3.csv", "w") as f:
  w = csv.writer(f)
  # will write each list line by line
  for i in data:
    w.writerow(i)
```

### Read CSV

```python
with open("test3.csv", "r") as f:
  read = csv.reader(f)
  for i in read:
    print(i)
```

## Binary Files

### Read Binary

```python
with open("test4.bin", 'wb') as f: # 'wb' write binary
  f.write(b"\x01\x02\x03\x04")
```

### Write Binary

```python
with open("test4.bin", "rb") as f:
  print(f.read())
```
