### Q. When to perform buffer read and write ?

- we use this when our file size is very-very big.
- when i have to perform a read operation on big files of size in GB's, TB's then what is the solution.
- solution is to read that file in chunks in small small bit's and pieces entire file.
- and this is where the buffer read and write comes into picture.


### Write Operation

```python
with open("test.txt", "wb") as f:
  file = io.BufferedWriter(f) # here buffer size is defauilt, but we can define the size of chunk
  file.write(b"this is my buffer write.\n")
  file.write(b"this is my second line that i am trying to write")
  # similarly like file objects we have to close file, but here we have to flush the object
  file.flush()
```

### Read Operation

```python
with open("test.txt", "rb") as f:
  file = io.BufferedReader(f)
  data = file.read()
  print(data)
```

- if we write `file.read(10)`, then it will read only 10 characters

```python
with open("test.txt", "rb") as f:
  file = io.BufferedReader(f)
  data = file.read(10)
  print(data)
```

- here we are controlling the output of chunk
