- `map()` is a function would like to perform given function on each elemnt provided

- it takes, `function` and `*iterables`

```python
def sq(x):
  return x**2
```

```python
map(sq, l) # it is going to return mapper object
```

- if we pass this mapper object to a list will get the output

```python
list(map(sq, l))
```

```python
l = [1,2,3,4,5,4]
```

Q. can you give summation of each elemnt present in list ?

- we can use an external function `reduce()`

```python
reduce(lambda x, y : x+y, l)
```

`1,2,3,4,5,4`

step1: `x=1, y=2 x+y -> 3`

list updates to `l = [3,3,4,5,4]`

step2: `x=3, y=3 x+y -> 6`

list updates to `l = [6,4,5,4]`

step3: `x=6, y=4 x+y -> 10`

list updates to `l = [10,5,4]`

step4: `x=10, y=5 x+y -> 15`

list updates to `l = [15,4]`

step5: `x=15, y=4 x+y -> 19`

output: `19`

Q. can you filter out only even number or odd number ?

filter function is same as `map() and reduce()` it takes function and iterables

```python
filter(lambda x : x % 2 ==0 , l )
```

```python
list(filter(lambda x : x % 2 ==0 , l ))
```

- based on condition where it is true it will filter the inputs

```python
list(filter(lambda x : x % 2 ==1 , l ))
```

Q. only get me either `+ve` or `-ve` elements ?

```python
l1 = [-2, 4,5,6,-3,-6,-7]
```

```python
list(filter(lambda x: x < 0, l1))
```

```python
list(filter(lambda x: x >= 0, l1))
```
