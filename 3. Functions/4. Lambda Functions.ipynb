```python
n = 3
p = 2
```

```python
def test(n, p):
  return n**p
```

```python
test(3,2)
```

```python
test(9,4)
```

diffrent way to write this function is *Lambda* function, ananymous function, function without name, it a ad-hoc function

Lambda Function to get Power

```python
# we are storing this function in a
a = lambda n,p : n**p
```

```python
a(3,2)
```

```python
a(9, 4)
```

Lambda Function to get sum of two numbers

```python
add = lambda x, y : x+y
```

```python
add(234, 34)
```

Lambda Function to convert celsius to fahrenheit

```python
# celsius to fahrenheit
c_to_f = lambda c : (9/5)*c+32
```

```python
c_to_f(34)
```

Lambda Function to find maximum of two numbers

```python
# lambda input1, input2: x(return this) if x>y else y(return this)

max_two = lambda x, y : x if x > y else y
```

```python
max_two(1, 2)
```

```python
max_two(4, 56)
```
