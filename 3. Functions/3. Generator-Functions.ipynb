What is Generator Functions ?

````python
range(1, 10) # returning only object
````

- it is generating data from 1 to 10(excluding 10), but it is not returning it, unless we call it inside a loop

````python
for i in range(1, 10):
  print(i)
````

````python
def test1(a):
  n = []
  for i in a:
    if type(i) == int:
      n.append(i)
  return n
````

- What is i have millions of elements in `a`, then my function will take infinite amount of time to give me output
- means as my list grows the complexity also grows

-> to solve this will use `generators`

- we can see in `test1`, data a is getting stored and program terminate only when it has created the list from a, if a has million or trillions of elements, then it becames a bottole nack
and also to remember it it get stored in system memory

- But if we see `range(1, 10)` object generated by genertor it doesn't have to store values in memory and it is less complex compare to `test1`


- if i have to create a generator like object i can use a function `yeild()`

```Fibonacci:

  0,1 ,1,2,3,4,7,11,18,29,34```

````python
def test_fib(n):
  a,b = 0,1
  for i in range(n):
    yield a
    a,b = b, a+b
````

````python
test_fib(10)
````

- it returned us a generator object

````python
for i in test_fib(10):
  print(i)
````
